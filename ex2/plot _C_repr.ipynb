## **Single Precision**

```python
import numpy as np
import matplotlib.pyplot as plt

# Carico i dati
N_max, err_norm, err_rev = np.loadtxt('output_float.txt', skiprows=2, unpack=True)

# Creo il plot
fig, ax = plt.subplots(figsize=(7,5))

ax.plot(N_max, err_norm, label='Normal ordering', color='red', lw=2)
ax.plot(N_max, err_rev, label='Reverse ordering', color='blue', lw=2)

ax.set_xlabel('N', fontweight='bold')
ax.set_ylabel(r'$\bf|S(N) - \pi^2/6|$')
ax.set_xscale('log')
ax.set_yscale('log')
ax.set_title('Single precision – Absolute error vs N', fontweight='bold', fontsize=13)
ax.grid(True, which='both', ls='--', lw=0.5)
ax.legend()

plt.tight_layout()
plt.show()
```
