```python
import numpy as np
import matplotlib.pyplot as plt
```

## **Single Precision**

```python

N_max, err_norm, err_rev = np.loadtxt('float.txt', skiprows=2, unpack=True)

fig, ax = plt.subplots(figsize=(7,5))

ax.plot(N_max, err_norm, label='Normal ordering', color='red', lw=2.5)
ax.plot(N_max, err_rev, label='Reverse ordering', color='blue', lw=2.5)
ax.vlines(np.sqrt(1.1e7), min(err_rev), max(err_rev), ls='--', lw=1.8, label=r'$\tilde{N}_N = 1/ \sqrt{\epsilon_{single}}$', colors='purple')
ax.vlines(1.1e7, min(err_rev), max(err_rev), ls='-.', lw=1.8, label=r'$\tilde{N}_R = 1/ \epsilon_{single}$', colors='green')

ax.set_xlabel('N', fontweight='bold')
ax.set_ylabel(r'$\mathbf{|S(N) - \pi^2/6|}$')
ax.set_xscale('log')
ax.set_yscale('log')
# ax.set_title('Single precision – Absolute error vs N', fontsize=15)
ax.grid(True, ls='--', lw=0.5)
ax.legend(bbox_to_anchor=(0.48, 0.35))

plt.tight_layout()
plt.savefig('plots/SinglePrec.png')
plt.show()
```

## **Double Precision**

```python
N_max, err_norm, err_rev = np.loadtxt('double.txt', skiprows=2, unpack=True)

fig, ax = plt.subplots(figsize=(7,5))

ax.plot(N_max, err_norm, label='Normal ordering', color='red', lw=2)
ax.plot(N_max, err_rev, label='Reverse ordering', color='blue', lw=2)
ax.vlines(np.sqrt(2.2e16), min(err_rev), max(err_rev), ls='--', lw=1.8, label=r'$\tilde{N}_N = 1/ \sqrt{\epsilon_{double}}$', colors='purple')
ax.vlines(2.2e16, min(err_rev), max(err_rev), ls='-.', lw=1.8, label=r'$\tilde{N}_R = 1/ \epsilon_{double}$', colors='green')


ax.set_xlabel('N', fontweight='bold')
ax.set_ylabel(r'$\mathbf{|S(N) - \pi^2/6|}$')
ax.set_xscale('log')
ax.set_yscale('log')
# ax.set_title('Double precision – Absolute error vs N', fontsize=15)
ax.grid(True, ls='--', lw=0.5)
ax.set_xlim(min(N_max)-0.4*min(N_max), max(N_max)+0.4*max(N_max))
ax.legend(bbox_to_anchor=(0.31, 0.35))

plt.tight_layout()
plt.savefig('plots/DoublePrec.png')
plt.show()
```
