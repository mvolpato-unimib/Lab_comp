A generic template (to be filled) for a QR function based on the previously introduced Householder reflector is:

```python
function QR(A):
    Q = identity nxn
    R = copy of A such that A is left untouched
    for i in range(...):
        Pp = householder( sub-block of R defined by i )
        P = manipulation of Pp to turn it into the appropriate nxn matrix
        R = P.R
        Q = Q.P
        
    return Q and R

## **Linear solver QR**

````python
A_test = np.array([[3, 2, -1],
              [2, -2,  4],
              [-1, 0.5, -1]])
b_test = np.array([1, -2, 0])

def QR_solver(A, b):
    """
    Solve a linear system using QR decomposition and Backward substitution.
    """
    Q,R = QR_dec(A)
    x_sol = back_subst(R, Q.T @ b, speak=True)
    return x_sol 

QR_solver(A_test, b_test)
    
````
