```python
'''
SINGLE PRECISION
'''
n1 = np.float32(1.2e34)

range1 = 15

for i in range(range1):
    n1 *= np.float32(2)
    print(f'{n1:.6e}')

print()

'''
DOUBLE PRECISION
'''
n1 = np.float64(1.2e304)

for i in range(range1):
    n1 *= np.float64(2)
    print(f'{n1:.6e}')
```

```python

'''
SINGLE PRECISION
'''
n1 = np.float32(1e-13)

range1 = 15

for i in range(range1):
    n1 /= np.float32(2)
    print(f'{n1:.6e}, {n1+np.float32(1):.6e}')

print()

'''
DOUBLE PRECISION
'''
n1 = np.float64(1e-13)

for i in range(range1):
    n1 /= np.float64(2)
    print(f'{n1:.6e}, {n1+np.float64(1):.6e}')
```
